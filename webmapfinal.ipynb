```python
import geopandas as gpd
import folium

# Load geospatial data for buildings and roads from GeoJSON files
mapad = gpd.read_file('abudhabi.geojson')
mapnoad = gpd.read_file('noad1.geojson')

# Set the latitude and longitude for the map's center location
lat, lng= 24.477544108154966, 54.418910616288436
map1_location = [lat, lng]

# Create a Folium map centered at the specified location with a starting zoom level
map = folium.Map(
    location=map1_location,
    zoom_start=12,
    tiles='CartoDB positron'
)


# Add buildings layer and roads layer to the map using GeoJSON data
folium.GeoJson(mapad.geometry, name = 'ad', color = '#19e3d9', opacity = 0.25).add_to(map)
folium.GeoJson(mapnoad.geometry, name = 'noad', color='#ad5603', opacity = 0.25).add_to(map)

# Display the interactive map
display(map)
map.save('admap.html')
```
